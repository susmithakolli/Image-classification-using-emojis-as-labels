```python
def image_file_to_np(ifile,resize=(64,64)):
    pil_img = Image.open(ifile)
    pil_img = pil_img.resize(resize,Image.ANTIALIAS)
    np_img = np.asarray(pil_img)
    pil_img.close()
    return np_img.copy()

pil_img = Image.open("C:\\Users\\susmitha kolli\\Desktop\\final_train\\happy1.20.jpg")
pil_img = pil_img.resize((64,64),Image.ANTIALIAS)
np_img = np.asarray(pil_img)

print(np_img)

def get_brightness(inp):
    w0 = 0.21
    w1 = 0.72
    w2 = 0.07
    r = inp[:,:,0]
    g = inp[:,:,1]
    b = inp[:,:,2]
    brightness = w0*r+w1*g+w2*b
    return brightness.copy()


def inp_to_hist_brightness(inp, bins = 10):
    z = get_brightness(inp)
    y = np.histogram(z, bins = bins, range = (0,255))[0]
    x = y.reshape(1,bins)
    return x.copy()

def inp_to_hist_bycolor(inp, bins = 10, channel=0):
    z = inp[:,:,channel]
    y = np.histogram(z, bins = bins, range = (0,255))[0]
    x = y.reshape(1,bins)
    return x.copy()

def np_img_and_bright_hist(inp,bins=10):
    a = np_img_to_hist(inp,bins=bins)
    b = inp_to_hist_brightness(inp,bins=bins)
        
    return np.hstack([a,b])

def np_img_to_hist(inp,bins=10):
    return np.hstack([ np.histogram(inp[:,:,x],bins=bins, range = (0,255))[0] for x in range(3)]).reshape(1,bins*3)


def plot_hist_all(i_np_hist):
    x_bins = np.arange(0,i_np_hist.shape[1])
    for x,h in zip(x_bins,i_np_hist.flatten()):
        
        if x<10:
            c='red'
        elif x<20:
            c='green'
        elif x<30:
            c='blue'
        else:
            c ="yellow"
        plt.bar(x,h,color=c)
    plt.show()
    
    
def plot_hist(i_np_hist):
    x_bins = np.arange(0,i_np_hist.shape[1])
    for x,h in zip(x_bins,i_np_hist.flatten()):
        
        if x<10:
            c='red'
        elif x<20:
            c='green'
        else:
            c='blue'
        plt.bar(x,h,color=c)
    plt.show()
    
    
def plot_single_hist(i_np_hist, color= "red"):
    x_bins = np.arange(0,i_np_hist.shape[1])
    for x,h in zip(x_bins,i_np_hist.flatten()):
        plt.bar(x,h,color=color)
    plt.show()
    
    
def batch_hist_from_df(idf,x_col='filename',y_col='target',batch_size=4,offset=4,resize=(64,64),bins=10, mode= "rgbh"):
    start_index = offset
    rows = idf.shape[0]
    if start_index >= rows:
        start_index = start_index%rows
    end_index = start_index + batch_size
    
    if end_index>rows:
        end_index = rows
    
    tdf = idf.iloc[start_index:end_index]
    _y = np.vstack(tdf[y_col])
    
    arrays = list(map(lambda x: image_file_to_np(x,resize=resize),tdf[x_col]))
    if mode == "red":
        _x = np.vstack(list(map(lambda x: inp_to_hist_bycolor(x,bins=bins, channel =0),arrays)))
    elif mode == "green":
        _x = np.vstack(list(map(lambda x: inp_to_hist_bycolor(x,bins=bins, channel =1),arrays)))
    elif mode == "blue":
        _x = np.vstack(list(map(lambda x: inp_to_hist_bycolor(x,bins=bins, channel =2),arrays)))
    elif mode == "brightness":
        _x = np.vstack(list(map(lambda x: inp_to_hist_brightness(x,bins=bins),arrays)))
    elif mode == "rgb":
        _x = np.vstack(list(map(lambda x: np_img_to_hist(x,bins=bins),arrays)))
    elif mode =="rgbh":
        _x = np.vstack(list(map(lambda x: np_img_and_bright_hist(x,bins=bins),arrays)))
   

    return _x.copy(),_y.copy()


        
```

```python
idf = train4

bins_per_channel = 128
# modes available are: red, blue , green, rgb, brightness
mode = "rgbh" 
mode_factor = {'red':1,'green':1,'blue':1,'rgb':3,'brightness':1, 'rgbh':4}

input_dims = bins_per_channel* mode_factor[mode]
output_classes = 6
iterations = 70
learning_rate=0.001
batch_size=10
resize_shape = (128,128)
x_col='file_full'
y_col='target'

#Model

model = Sequential()

model.add(Dense(units=16, activation='relu', input_dim=input_dims))
model.add(BatchNormalization())
model.add(Dense(units=16, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=16, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=output_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.SGD(lr=learning_rate, momentum=0.9, nesterov=True))

# Iterations

rows = image_df.shape[0]
batches = int(np.ceil(rows/batch_size))

for iterx in range(iterations):
    for batch in range(batches):
        offset = batch*batch_size
        xb,yb = batch_hist_from_df(idf=idf,x_col=x_col,y_col=y_col,
                           batch_size=batch_size,offset=offset,
                            resize=resize_shape,bins=bins_per_channel,mode ="rgbh")
        
        loss = model.train_on_batch(xb, yb)
        print("Iter",iterx,"/",iterations,"Batch",batch,"/",batches,"Loss:",loss)
```

```python
test_rows = train4.shape[0]
tbatches = int(np.ceil(test_rows/batch_size))
general_accuracy = []
ground_truths =[]
class_preds = []


for batch in range(tbatches):
    offset = batch*batch_size
    xb,yb = batch_hist_from_df(idf=train4,x_col=x_col,y_col=y_col,
                           batch_size=batch_size,offset=offset,
                            resize=resize_shape,bins=bins_per_channel)
    print(xb.shape, yb.shape)
    classes = model.predict(xb)
    print(classes)
    loss = model.evaluate(xb, yb,verbose=0)
    
    class_pred = np.argmax(classes,1)
    print(class_pred)
    print("Batch",batch,"/",tbatches,"Loss:",loss)
    ground_truth = np.argmax(yb,1)
    print(ground_truth)
    
    accuracy = sum(np.equal(class_pred,ground_truth))/batch_size
    general_accuracy.append(accuracy)   
    
    ground_truths.append(ground_truth)
    class_preds.append(class_pred)
    
    
#print(xb.shape, yb.shape)


print("Total accuracy",np.mean(general_accuracy))
```

```python
import itertools


def plot_confusion_matrix(cm, classesNames,printCM=False,normalize=False, title='Confusion matrix', cmap=plt.cm.Blues,figsize=(12,6),dpi=80):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    cm is the confusion matrix
    """
    plt.figure(figsize=figsize,dpi=dpi)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classesNames))
    plt.xticks(tick_marks, classesNames, rotation=45)
    plt.yticks(tick_marks, classesNames)

    if normalize:
        cm = np.round(cm.astype('float') / cm.sum(axis=1)[:, np.newaxis],2)
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    if printCM==True:
        print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```
